```python
import matplotlib.pyplot as plt
labels  = ['apple', 'banana', 'cherry', 'date', 'elderberry']
sizes = [2, 3, 5, 7, 11]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("fruit pie chart")
plt.xlabel("categories")
plt.ylabel("Values")
plt.grid(True)
plt.legend(["GREAT WORK"])

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04'],
    'Sales': [1000, 1500, 120, 180]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Line Plot
df.plot(x='Date', y='Sales', kind='line', title='Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(0, 5, 100)
y1 = np.sin(x)       
y2 = np.cos(x)       
y3 = np.sin(x) * 0.5 

plt.plot(x, y1, label='sin(x)', color='pink', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='blue', linestyle='--', linewidth=2)
plt.plot(x, y3, label='0.5*sin(x)', color='green', linestyle=':', linewidth=2)


plt.xlabel("X-axis", fontsize=18)
plt.ylabel("Y-axis", fontsize=18)

plt.savefig("plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)  
plt.plot([1, 2, 3], [4, 5, 6])
plt.title("Plot 1")

plt.subplot(1, 2, 2)  
plt.bar([1, 2, 3], [5, 7, 3])
plt.title("Plot 2")

plt.tight_layout()  
plt.savefig("plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(18, 8))
plt.plot(x, y, label='sin(x)')
plt.title('Sine Wave (ggplot style)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.savefig("plot.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, color='black', linewidth=10)

plt.title("Title", fontsize=24, fontweight='bold')  
plt.xlabel("X-axis", fontsize=10)
plt.ylabel("Y-axis", fontsize=18)


plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("plot.png")
plt.show()
```
